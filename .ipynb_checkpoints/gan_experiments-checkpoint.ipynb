```python
import torch

def psnr(errs):
    return -10 * torch.log(errs) / torch.log(10. * torch.ones(errs.size()))
```

```python
import torch
import torch.nn.functional as F

class Model(object):
    '''
        Builds Net object and Stats object 
    '''
    def __init__(self, net, device='cpu'):
        self.device = device
        self.net = net.to(device)
        self.criterion = torch.nn.MSELoss(reduction='none').to(device)
        self.optimizer = torch.optim.Adam(net.parameters())
        
        self.stats = SRStats()
        
        self.metric = psnr
    
    @staticmethod
    def load(self):
        pass
    
    def save(self, output_filename):
        torch.save(
            {
                'net': self.net,
                'optimizer': self.optimizer,
            }, 
            output_filename
#             join(output_directory, "net_and_optimizer.pt")
        )
         
    def train_on_batch(self, batch_lr, batch_hr, train=True, label='train'):
        self.optimizer.zero_grad()
        
#         batch_lr = lr.to(self.device)
#         batch_hr = hr.to(self.device)
        
        predicted_hr = self.net.forward(batch_lr)

        raw_loss = (predicted_hr - batch_hr) ** 2
        average_picture_loss = F.avg_pool3d(raw_loss, raw_loss.shape[1:]).flatten()
        loss = torch.mean(average_picture_loss)
        
        if train:
            loss.backward()
            self.optimizer.step()
        
        picture_errs = average_picture_loss.detach().cpu()
        metrics = self.metric(picture_errs)

        self.stats.add_set_stats(label, picture_errs, metrics)
    
    def evaluate_on_batch(self, lr, hr):     
        self.train_on_batch(lr, hr, False, 'test')

    def plot(self, output_filename):
        return self.stats.plot(output_filename)
    
    def forward(self, data):
        return self.net.forward(data)
    
    def predict(self, data):
        return self.net.predict(data)
    
```
